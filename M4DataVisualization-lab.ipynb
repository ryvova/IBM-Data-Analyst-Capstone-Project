## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'DatabaseDesireNextYear'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
QUERY = """
SELECT ConvertedComp
FROM master
"""

df = pd.read_sql_query(QUERY, conn)
df.hist()
```

### Box Plots


Plot a box plot of `Age.`


```python
QUERY = """
SELECT Age
FROM master
"""

df = pd.read_sql_query(QUERY, conn)
df.boxplot()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
QUERY = """
SELECT Age, WorkWeekHrs
FROM master
"""

df = pd.read_sql_query(QUERY, conn)
df.plot.scatter('Age', 'WorkWeekHrs', title = 'WorkWeekHrs by Age')
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
import seaborn as sns

QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age
FROM master
"""

df = pd.read_sql_query(QUERY, conn)
sns.scatterplot(y = 'WorkWeekHrs', x = 'CodeRevHrs', size = 'Age', hue = 'Age', alpha = 0.5, data = df, title = 'CodeRevHrs And WorkWeekHrs by Age', sizes = (10 , 500))

```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
QUERY = """
SELECT DatabaseDesireNextYear AS DB, Count(*) AS Count
FROM DatabaseDesireNextYear
GROUP BY DB
ORDER BY Count DESC
LIMIT 5
"""

df = pd.read_sql_query(QUERY, conn)
df.groupby(['DB']).sum().plot(kind='pie', y='Count', ylabel = '', autopct='%1.0f %%', figsize = (4, 4), title = 'TOP 5 Databases Desired Next Year').legend(bbox_to_anchor = (1.2, 1.15), loc = 'upper left')
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
Query = '''
SELECT WorkWeekHrs, CodeRevHrs, Age 
FROM master 
WHERE Age BETWEEN 30 AND 35
'''

df = pd.read_sql_query(Query,conn).groupby('Age').median()
df[['WorkWeekHrs', 'CodeRevHrs']].plot(kind ='bar', stacked = True, ylabel = 'Hours', title = 'Median WorkWeekHrs And CodeRevHrs by Age', legend = True).legend(bbox_to_anchor = (1.1, 1), loc = 'upper left')
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
Query = '''
SELECT ConvertedComp, Age 
FROM master 
WHERE Age BETWEEN 45 AND 60
'''

df = pd.read_sql_query(Query,conn).groupby('Age').median()
df['ConvertedComp'].plot(kind ='line', title = 'Median ConvertedComp by Age', legend = True).legend(bbox_to_anchor = (1.1, 1), loc = 'upper left')
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
Query = '''
SELECT MainBranch, COUNT(MainBranch) AS Count
FROM master 
GROUP BY MainBranch
'''

df = pd.read_sql_query(Query,conn)
df.set_index('MainBranch', inplace=True)
df.plot(kind = 'barh', ylabel = 'MainBranch', title = 'Frequency of Mainbranch')
```
